```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
os.chdir ("D:\PaAC\Astro Sonification\Assignment 3")
image_data = np.load('images.npy')
```

```python
# Selecting a high-resolution image with indox 0
selected_image = image_data[0]

# Checking the dimensions of the selected image
print(f"Selected image shape: {selected_image.shape}")
```

```python
# Saveing RGB values to a CSV file
height, width, channels = selected_image.shape

with open('high_res_galaxy_image_rgb.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    # Write the header row
    writer.writerow(['Row', 'Column', 'Red', 'Green', 'Blue'])
    
    # Loop through each pixel and extract its RGB values
    for row in range(height):
        for col in range(width):
            r, g, b = selected_image[row, col]
            writer.writerow([row, col, r, g, b])

print("RGB values saved to 'high_res_galaxy_image_rgb.csv'.")
```

```python
# Step 4: Visualize the image using matplotlib
plt.imshow(selected_image)
plt.title('High-Resolution Galaxy Image')
plt.show()
```

```python
# Visualizing pixel intensity like the Red channel intensity
red_channel = selected_image[:, :, 0]  # Extracting the red channel

plt.imshow(red_channel, cmap='Reds')
plt.title('Red Channel Intensity')
plt.colorbar()
plt.show()
```
